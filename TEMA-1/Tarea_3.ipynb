1. Find all of the numbers from 1-1000 that are divisible by 7.

2. Find all of the numbers from 1-1000 that have a 3 in them.


```python
def detector_3(number):
    number = str(number)
    x = len(number)
    if x == 1:
        if number[0] == '3':
            return number
    if x == 2:
        if (number[0] == '3') or (number[1] == '3'):
            return number
    
    if x == 3:
        if (number[0] == '3') or (number[1] == '3') or (number[2] == '3'):
            return number
            
```

```python
z = np.arange(1001)
z
```

```python
#num_w3 = list(map(detector_3, z))
num_w3 = [detector_3(j) for j in z]
num_w3 = [i for i in num_w3 if i is not None]
num_w3
```

3. Count the number of spaces in a string.


```python
def blank_(char):
    count = 0
    if char == " ": 
        count = 1
        return count
    else: 
        return count
        
```

```python
txt = 'Esto es un ejemplo '
#cuenta = np.sum(np.array(list(map(blank_, txt))))
cuenta = np.sum(np.array([blank_(i) for i in txt]))
cuenta
```

4. Remove all of the vowels in a string [make a list of the non-vowels].


```python
test = "GeeksforGeeks - A Computer Science Portal for Geeks"

vowels = ('a', 'e', 'i', 'o', 'u') 
f2 = lambda x: None if (x.lower() in vowels) or (x == ' ') else x
c1 = [f2(j) for j in test]
c1 = [i for i in c1 if i is not None]
print(c1)

```

5. Find all of the words in a string that are less than 4 letters.


```python
old_string = 'hol como estas'
new_string = ' '.join([w for w in old_string.split() if len(w)<4])
new_string
```

6. Use a dictionary comprehension to count the length of each word in a sentence.
